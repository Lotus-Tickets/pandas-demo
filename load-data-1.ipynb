```python
import pandas as pd
import json
```

### Load `inventory_response_data` json

```python
with open('inventory_response_1.json') as f:
    data = json.load(f)

print("num rows: ", len(data['rows']))

inventory_df = pd.DataFrame(data['rows'])

print("inventory_df shape: ", inventory_df.shape)

inventory_df.head(5)
```

```python
list(inventory_df.columns)[:8]
```

```python
dtypes = inventory_df.dtypes
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(dtypes[:20])
```

### unravel objects within a field: e.g. `event`

```python
event_df = inventory_df['event'].apply(pd.Series)
print(event_df.shape)
event_df.head(5)
```

```python
event_df.dtypes
```

```python
event_venue_df = event_df['venue'].apply(pd.Series)

print(event_venue_df.shape)
event_venue_df.head(5)
```

### Build Full Data

```python
event_df.columns = ["event_" + col for col in event_df.columns]
event_venue_df.columns = ["event_venue_" + col for col in event_venue_df.columns]
```

```python
df = pd.concat([inventory_df, event_df, event_venue_df], axis=1 )
```
